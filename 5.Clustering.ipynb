```python
from sklearn.datasets import make_blobs
from sklearn.cluster import KMeans, AgglomerativeClustering
from sklearn.mixture import GaussianMixture
import matplotlib.pyplot as plt

X, _ = make_blobs(n_samples=300, centers=3, random_state=42)

kmeans = KMeans(n_clusters=3)
k_labels = kmeans.fit_predict(X)

gmm = GaussianMixture(n_components=3)
g_labels = gmm.fit_predict(X)

hc = AgglomerativeClustering(n_clusters=3)
h_labels = hc.fit_predict(X)

fig, axs = plt.subplots(1, 3, figsize=(12, 4))
axs[0].scatter(X[:, 0], X[:, 1], c=k_labels)
axs[0].set_title("K-Means")
axs[1].scatter(X[:, 0], X[:, 1], c=g_labels)
axs[1].set_title("Gaussian Mixture")
axs[2].scatter(X[:, 0], X[:, 1], c=h_labels)
axs[2].set_title("Hierarchical Clustering")
plt.show()
```
